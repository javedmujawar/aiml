# Q.1 Write a python program to demonstrate the use of LIST with appropriate method

```python
# Creating a list
fruits = ['apple', 'banana', 'orange', 'grape']

# 1. Accessing elements of the list
print("First fruit:", fruits[0])
print("Last fruit:", fruits[-1])

# 2. Adding elements to the list
fruits.append('kiwi')
print("Fruits after adding 'kiwi':", fruits)

# 3. Removing elements from the list
removed_fruit = fruits.pop(2)
print("Removed fruit:", removed_fruit)
print("Fruits after removing the third fruit:", fruits)

# 4. Modifying elements of the list
fruits[1] = 'pineapple'
print("Fruits after replacing 'banana' with 'pineapple':", fruits)

# 5. Sorting the list
fruits.sort()
print("Fruits after sorting:", fruits)

# 6. Reversing the list
fruits.reverse()
print("Fruits after reversing:", fruits)

# 7. Finding the index of an element
index = fruits.index('apple')
print("Index of 'apple':", index)

# 8. Counting occurrences of an element
count = fruits.count('orange')
print("Number of 'orange' in the list:", count)

# 9. Slicing the list
print("First three fruits:", fruits[:3])

# 10. Clearing the list
fruits.clear()
print("Fruits list after clearing:", fruits)
```

# Q.2 Write a python program to demonstrate the use of SET with appropriate method

```python
# Creating a set
fruits = {"apple", "banana", "orange", "grape", "banana"}

# Displaying the original set
print("Original set of fruits:", fruits)

# Adding an element to the set
fruits.add("pineapple")
print("After adding 'pineapple':", fruits)

# Removing an element from the set
fruits.remove("banana")
print("After removing 'banana':", fruits)

# Checking if an element exists in the set
print("Is 'apple' in the set?", "apple" in fruits)

# Iterating over the set
print("Fruits in the set:")
for fruit in fruits:
    print(fruit)

# Getting the length of the set
print("Number of fruits in the set:", len(fruits))

# Creating another set
more_fruits = {"kiwi", "melon", "orange", "grape"}

# Performing set operations
print("Intersection of the two sets:", fruits.intersection(more_fruits))
print("Union of the two sets:", fruits.union(more_fruits))
print("Difference between the two sets:", fruits.difference(more_fruits))
print("Symmetric difference between the two sets:", fruits.symmetric_difference(more_fruits))

# Clearing the set
fruits.clear()
print("After clearing the set:", fruits)
```

# Q.3 Write a python program to demonstrate the use of DICTIONARY with appropriate method

```python
# Creating a dictionary
student_scores = {
    "Alice": 85,
    "Bob": 72,
    "Charlie": 90,
    "David": 68,
    "Eve": 95
}

# Accessing values using keys
print("Alice's score:", student_scores["Alice"])

# Adding a new key-value pair
student_scores["Frank"] = 80
print("Added Frank's score:", student_scores)

# Modifying a value
student_scores["Bob"] = 75
print("Bob's updated score:", student_scores)

# Removing a key-value pair
del student_scores["David"]
print("Dictionary after removing David:", student_scores)

# Checking if a key exists
if "Alice" in student_scores:
    print("Alice's score:", student_scores["Alice"])

# Iterating through keys
print("Students and their scores:")
for student in student_scores:
    print(student, ":", student_scores[student])

# Iterating through keys and values
print("Students and their scores:")
for student, score in student_scores.items():
    print(student, ":", score)

# Getting the length of the dictionary
print("Number of students:", len(student_scores))

# Clearing the dictionary
student_scores.clear()
print("Cleared dictionary:", student_scores)
```

# Q.4 Write a python program to demonstrate the use of TUPLE with appropriate method

```python
# Define a tuple
my_tuple = (1, 2, 3, 4, 5)

# Accessing elements of a tuple
print("Tuple elements:", my_tuple)
print("First element:", my_tuple[0])
print("Last element:", my_tuple[-1])

# Tuple slicing
print("Sliced tuple:", my_tuple[1:4])

# Tuple length
print("Length of tuple:", len(my_tuple))

# Counting occurrences of an element in a tuple
print("Count of '3' in tuple:", my_tuple.count(3))

# Finding index of an element in a tuple
print("Index of '4' in tuple:", my_tuple.index(4))

# Converting a tuple to a list
tuple_list = list(my_tuple)
print("Tuple converted to list:", tuple_list)

# Concatenating tuples
another_tuple = (6, 7, 8)
concatenated_tuple = my_tuple + another_tuple
print("Concatenated tuple:", concatenated_tuple)
```

# Q.5 Write a python program to demonstrate the use of NUMPY library

```python
import numpy as np

# Creating a 1D array
arr1d = np.array([1, 2, 3, 4, 5])
print("1D Array:")
print(arr1d)

# Creating a 2D array
arr2d = np.array([[1, 2, 3], [4, 5, 6]])
print("\n2D Array:")
print(arr2d)

# Array shape
print("\nShape of 1D Array:", arr1d.shape)
print("Shape of 2D Array:", arr2d.shape)

# Accessing elements
print("\nAccessing elements:")
print("First element of 1D Array:", arr1d[0])
print("Element at row 1, column 2 of 2D Array:", arr2d[0, 1])

# Array operations
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])

print("\nArray operations:")
print("Sum of arrays:")
print(arr1 + arr2)
print("Element-wise multiplication of arrays:")
print(arr1 * arr2)
print("Matrix multiplication of arrays:")
print(np.dot(arr1, arr2))
```

# Q.6 write a python prg to demonstrate the use of function each function should contain one looping type

```python
def while_loop_example():
    # Function to demonstrate while loop
    num = 1
    while num <= 5:
        print(num)
        num += 1

def for_loop_example():
    # Function to demonstrate for loop
    for i in range(1, 6):
        print(i)

def nested_loop_example():
    # Function to demonstrate nested loop
    for i in range(1, 4):
        for j in range(1, 4):
            print(f'({i}, {j})')

def do_while_loop_example():
    # Function to demonstrate do-while loop (Python doesn't have a built-in do-while loop,
    # but similar behavior can be achieved with a while loop)
    num = 1
    while True:
        print(num)
        num += 1
        if num > 5:
            break

# Call each function to demonstrate the loop type
print("While Loop Example:")
while_loop_example()
print("\nFor Loop Example:")
for_loop_example()
print("\nNested Loop Example:")
nested_loop_example()
print("\nDo-While Loop Example:")
do_while_loop_example()
```

# Q.7 write a python prg to demonstrate the use of various Date function

```python
import datetime

def current_date():
    # Function to get the current date
    current_date = datetime.date.today()
    print("Current Date:", current_date)

def date_operations():
    # Function to demonstrate date operations
    today = datetime.date.today()
    future_date = today + datetime.timedelta(days=30)
    print("Today's Date:", today)
    print("Date after 30 days:", future_date)

def date_formatting():
    # Function to demonstrate date formatting
    today = datetime.date.today()
    formatted_date = today.strftime("%d-%m-%Y")
    print("Formatted Date:", formatted_date)

def custom_date():
    # Function to create a custom date
    custom_date = datetime.date(2024, 4, 22)
    print("Custom Date:", custom_date)

# Call each function to demonstrate various date functions
print("Current Date:")
current_date()
print("\nDate Operations:")
date_operations()
print("\nDate Formatting:")
date_formatting()
print("\nCustom Date:")
custom_date()
```
